```python
#import numpy as np
#import matplotlib.pyplot as plt

#-- Generate Data -----------------------------------------
# Using linspace so that the endpoint of 360 is included...
azimuths = np.radians(np.linspace(0, 360, 72))
zeniths = np.arange(0, 70, 1)

r, theta = np.meshgrid(zeniths, azimuths)

avalues = np.array([list(np.random.vonmises(0,3,azimuths.size)),list(np.random.random(zeniths.size))])
values = np.random.randn(azimuths.size,zeniths.size)
```
